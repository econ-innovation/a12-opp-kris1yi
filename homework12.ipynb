## 2. 第二张表为论文的摘要信息，表格的主键为ut论文入藏号，包括入藏号、摘要 

```python
table2_columns = ['UT', 'AB']
table2 = df[table2_columns]
table2
```

## 3. 第三张表为论文的题目，表格的主键为ut论文入藏号，包括入藏号、标题

## 4. 第四张表为论文的作者，表格的主键为ut论文入藏号与作者，包括入藏号、作者全名、作者的family name、givenname、作者顺序;

```python
table4_columns = ['UT', 'AF', 'C1']
table4 = df[table4_columns].copy()
table4
```

```python
table4_new = pd.DataFrame({'UT':[], 'author':[], 'family name':[], 'given name':[], 'author order':[],
                           'address':[]})

for index, row in table4.iterrows():
    all_authors = row['AF'].split('; ')
    for j, author in enumerate(all_authors):
        table4_new.loc[len(table4_new.index)] = [
            row['UT'], author, author.split(', ')[0], author.split(', ')[1], j+1, row['C1']
        ]

table4_new
```

## 5. 第五张表为论文的作者与单位，表格的主键为ut论文入藏号与作者，包括入藏号、作者全名、作者顺序、单位affiliation地址全文、单位署名顺序;

```python
import re

reg = re.compile("\\[[^\\]]+\\][^;]+")
reg.findall("[Gopinath, Gita; Stein, Jeremy C.] Harvard Univ, Cambridge, MA 02138 USA; [Gopinath, Gita; Stein, Jeremy C.] NBER, Cambridge, MA 02138 USA")

reg2 = re.compile("\\[[^\\]]+\\]")
reg2.findall('[Gopinath, Gita; Stein, Jeremy C.] Harvard Univ, Cambridge, MA 02138 USA')
def extract_author_affiliations(address):
    try:
        address_list = reg.findall(address)
    except TypeError:
        return {}
    au2add = {}
    for add in address_list:
        authors = reg2.findall(add)[0]
        for aut in authors[1:-1].split('; '):
            au2add[aut] = add.split(authors)[1:]
    
    return au2add


table5 = pd.DataFrame({'UT':[], 'author':[], 'family name':[], 'given name':[], 'author order':[],
                       'affiliation':[]})

for index, row in table4_new.iterrows():
    aut_dict = extract_author_affiliations(row['address'])
    if len(aut_dict) == 0:
        table5.loc[len(table5.index)] = [
            row['UT'], row['author'], row['family name'], row['given name'], row['author order'],
            row['address']
        ]
        continue
        
    table5.loc[len(table5.index)] = [
        row['UT'], row['author'], row['family name'], row['given name'], row['author order'],
        aut_dict.get(row['author'][0][1:])
    ]

table5
            
```

## 第六张表为论文的参考文献信息，表格的主键为ut与参考文献条目，包括ut与参考文献两列
